```python
# Cinderella story defined in sample.txt
with open('../documentation_mining/data/summarized_chunks_newtonsoft.txt', 'r') as file:
    text = file.read()

print(text[:100])
```

```python
from raptor import RetrievalAugmentation 
```

```python
RA = RetrievalAugmentation()

# construct the tree
RA.add_documents(text)
```

```python
question = "In newtonsoft, what is the purpose of the 'JsonConvert.DeserializeObject' method?"

answer = RA.answer_question(question=question)

print("Answer: ", answer)
```

```python
# Save the tree by calling RA.save("path/to/save")
SAVE_PATH = "my_try/newtonsoft_doc"
RA.save(SAVE_PATH)
```

```python
# load back the tree by passing it into RetrievalAugmentation

RA = RetrievalAugmentation(tree=SAVE_PATH)

answer = RA.answer_question(question=question)
print("Answer: ", answer)
```
